```python
import seaborn as sns
tips=sns.load_dataset('tips')
tips.head(n=3)
```

```python
print(tips.shape)
```

## We will be doing data Visulaliztion on this data frame

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.countplot(x='time',data=tips)
```

```python
sns.boxplot(x="day",y="total_bill",data=tips)
```

## Like say on thursday minimum bill is around 7 and maximum is around 30 and most of the data is between 10 to 20 and the line inside the box is the mean of the box 

## The dot above the box are known as outliers i.e say on friday the total a single table has a bill of 42

```python
sns.boxplot(x="day",y="total_bill",data=tips,hue='smoker')
```

## We can see that on sunday there are lot of smokers come to the restranunt

```python
sns.boxplot(x="day",y="total_bill",data=tips,hue='sex')
```

## Almost all day male are spending More money

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'],kde=False)
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=40)
```

```python
sns.kdeplot(tips['total_bill'])
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
```

## As the total bill is increasing the tip is also getting increased


```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue='sex')
```

## Now we are getting different pair plot for different sex 

```python
tips.head()
```

```python
tip_corr=tips.corr()
tip_corr
```

## It has taken all the numeric data and find all the relation between the pairs
## 1.000000 it means that both are increasing linearly at the same rate
## so generally they ranges from -1 to 1

```python
sns.heatmap(tip_corr)
```

```python
sns.heatmap(tip_corr,annot=True)
```
