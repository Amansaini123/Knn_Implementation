# Gradient descent Algorithm

```python
import numpy as np
import matplotlib.pyplot as plt

X=np.arange(10)
Y=(X-5)**2

print(X,Y)
```

```python
plt.style.use("seaborn")
plt.plot(X,Y)
plt.ylabel("F(x)")
plt.xlabel("X")
plt.show()
```

### gradient descent implementation

```python
x=0
learning_rate=0.1
error=[]

plt.plot(X,Y)
# say max iteration is 50
for i in range(50):
    grad=2*(x-5)
# grad is a differitial of y wrt to x
    x=x-(learning_rate*grad)
    y=(x-5)**2
    error.append(y)
    plt.scatter(x,y)
  
plt.plot(error)
```

# Relation between acidity and density of Milk using linear regression

## Linear_X.csv and Linear_Y.csv are the datasets used here

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


# x -->Acidity of Milk
# y -->Density of Milk




# pandas have a function of read_csv 
dfx = pd.read_csv('./Linear_Regression_Data/Linear_X.csv')
dfy = pd.read_csv('./Linear_Regression_Data/Linear_Y.csv')



# .values convert csv file to array
x = dfx.values
y = dfy.values




# -1 means we don't know how many rows and we are not specifying the cols means take all the cols
# Basically a way for flatten of all the rows and cols into a single array
x=x.reshape((-1,))
y=y.reshape((-1,))





# We are doing Normalization i.e we are shifting every point by same mean value so that they can converge early i.e 
# shifting all point toward the origin making mean to be 0 and standard deviation 1 i.e most of the entry are on the 
# scale 0 to 1 and their mean is 0
plt.scatter(x,y,color='red')
X = (x-x.mean())/x.std()
Y= y
plt.scatter(X,Y,color='green')
plt.title("Relation of Acidity and Density of Milk")
plt.show()




def hypothesis(theta,x):
    return theta[0]+theta[1]*x




def error(X,Y,theta):
    total_error=0
    m=X.shape[0]
    for i in range(m):
        total_error=total_error+(Y[i]-hypothesis(theta,X[i]))**2
    return 0.5*total_error




def gradient(Y,X,theta):
    grad=np.array([0.0,0.0])
    m=X.shape[0]
    for i in range(m):
        grad[0]+=-1*(Y[i]-hypothesis(theta,X[i]))
        grad[1]+=-1*(Y[i]-hypothesis(theta,X[i]))*X[i]
    return grad



def gradientdescent(X,Y,learn_rate,max_itr):
    grad=  np.array([0.0,0.0])
    theta= np.array([0.0,0.0])
    
    err=[]
    for i in range(max_itr):
        grad=gradient(Y,X,theta)
        curr_err=error(X,Y,theta)
#         print(curr_err)
        theta[0]=theta[0]-learn_rate*grad[0]
        theta[1]=theta[1]-learn_rate*grad[1]
        err.append(curr_err)
    return theta,err





```

```python
theta,total_error=gradientdescent(X,Y,learn_rate=0.001,max_itr=100)
print(theta[0],theta[1])
plt.plot(total_error)
plt.ylabel("error")
plt.xlabel("number of iterations")
plt.title("Relation between error and nuber of itr")
plt.show()

# We can vary number of iterations parameter and see their effect by the below graph graph i.e from here it is 
# evident that we need atlmost 40 iterations 
```

```python
# To show the graphical relation between values of X and Y and predicted Values of X and Y 
plt.scatter(X,Y,color='g')
plt.plot(X,hypothesis(theta,X),color='r')
```
